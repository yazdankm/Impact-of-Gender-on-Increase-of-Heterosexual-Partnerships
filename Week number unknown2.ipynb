```python
#Q18 
# using the cumulative distribution function (CDF) of the normal distribution.
# The CDF gives the probability that a random variable is less than or equal to a specified value.
# But the problem is the mean and the std could be calculated only for 
# 1.smoking mothers or 2. non smoking mothers or 3. all mothers


# first based on the mean and std for all mothers


import scipy.stats as stats

# Parameters
mean_birth_weight = 3.312857142857143
std_birth_weight = 0.603895022836148

# Specify the value (4.2 kg)
value = 4.2

# Calculate the probability using the CDF
probability_less_than_4_2 = stats.norm.cdf(value, loc=mean_birth_weight, scale=std_birth_weight)

print(f"Probability that birth weight is less than 4.2 kg: {probability_less_than_4_2:.4f}")
```

```python
#Q19 and Q20

import scipy.stats as stats

# Assuming 'df' is your DataFrame and 'Length' is the column for baby length
non_smoking_length_data = bw_df[bw_df['smoker'] == 0]['Length']

# Perform the Shapiro-Wilk test
statistic, p_value = stats.shapiro(non_smoking_length_data)

print(f"Shapiro-Wilk test statistic: {statistic}")
print(f"P-value: {p_value}")

# Check the significance level (e.g., 0.05)
alpha = 0.05
if p_value > alpha:
    print("The length data for non-smoking mothers appears to be normally distributed.")
else:
    print("The length data for non-smoking mothers does not appear to be normally distributed.")
```

```python
#Q21

mean_length_non_smoking = bw_df[bw_df['smoker'] == 0]['Length'].mean()
std_length_non_smoking = bw_df[bw_df['smoker'] == 0]['Length'].std()

# Specify the value (48.5 cm)
length_value = 48.5

# Calculate the z-score
z_score = (length_value - mean_length_non_smoking) / std_length_non_smoking

print(f"Standard Score (z-score) for a length of 48.5 cm in non-smoking mothers: {z_score:.4f}")
```

```python
#Q22

import scipy.stats as stats

# Assuming 'df' is your DataFrame and 'Length' is the column for baby length
mean_length_non_smoking = bw_df[bw_df['smoker'] == 0]['Length'].mean()
std_length_non_smoking = bw_df[bw_df['smoker'] == 0]['Length'].std()

# Specify the value (55 cm)
length_value = 55

# Calculate the z-score
z_score = (length_value - mean_length_non_smoking) / std_length_non_smoking

# Use the cumulative distribution function (CDF) to get the probability
probability_more_than_55cm = 1 - stats.norm.cdf(z_score)

print(f"Probability that the length of a baby for non-smoking mothers will be more than 55 cm: {probability_more_than_55cm:.4f}")
```
